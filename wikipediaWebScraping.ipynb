```python
import requests

import bs4

import pandas as pd
```

```python
OurData=requests.get('https://en.wikipedia.org/wiki/Artificial_intelligence')
```

```python
OurData.text
```

```python
bsp=bs4.BeautifulSoup(OurData.text,'xml')
```

```python
bsp
```

```python
title=bsp.select('title')
```

```python
title
```

```python
title[0].getText()
```

```python
title_class=bsp.select('h3')
```

```python
title_class
```

```python
for i in title_class:
   print(i.text)
```

```python
title_class[5].getText()
```
